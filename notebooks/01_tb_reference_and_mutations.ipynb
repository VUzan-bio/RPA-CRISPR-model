```python
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord

record = SeqIO.read("data/sequences/H37Rv_NC_000962.3.fasta", "fasta")
genome = record.seq

# Coordonnées relevées dans GenBank
start_rpoB, end_rpoB = 3648208, 3651726      # complement(...)
start_katG, end_katG = 2255422, 2257644      # +
start_inhA, end_inhA = 2736522, 2737331      # complement(...)

regions = {
    "rpoB_region": {
        "start": start_rpoB - 200,
        "end":   end_rpoB   + 200
    },
    "katG_region": {
        "start": start_katG - 200,
        "end":   end_katG   + 200
    },
    "inhA_promoter": {
        "start": start_inhA - 200,
        "end":   start_inhA + 50
    }
}

# Extraction et écriture des FASTA
for name, coords in regions.items():
    s = coords["start"] - 1   # 0-based
    e = coords["end"]
    subseq = genome[s:e]

    rec = SeqRecord(
        subseq,
        id=name,
        description=f"{name} {coords['start']}..{coords['end']} on NC_000962.3"
    )
    SeqIO.write(rec, f"data/sequences/{name}.fasta", "fasta")
    print(f"✓ Saved {name}.fasta ({len(subseq)} bp)")
```
